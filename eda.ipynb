```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
data = {
'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female'],
'Education_Level': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor', 'High School', 'PhD'],
'Age': [22, 27, 25, 29, 32, 30, 35],
'Salary': [2000, 3000, 2500, 4000, 3200, 2800, 4500],
'Satisfaction_Level': ['Low', 'Medium', 'High', 'Medium', 'High', 'Low', 'High']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
plt.hist(df['Age'],bins=5,color='blue')
plt.title("XYZ")
plt.xlabel('AGE')
plt.ylabel('YZ')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram for Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Age'], color='skyblue')
plt.title('Box Plot for Age')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df['Age'], shade=True, color='skyblue')
plt.title('Density Plot for Age')
plt.show()
```
